# **Unit 0: Robotnik Summit XL platform for research and surveillance**

* The SUMMIT XL has **skid-steering kinematics** based on 4 high power motorwheels. 

* This means that it turns by closely controlling the difference in speed turning  of each wheel.

* This removes the use of any steering mechanism, which make it highly robust.

## **1. Demo Move Around**

### **1. Publishing** topic **`/cmd_vel`** directly.

This robot need that you publish the Twist message continuously, otherwise it will stop moving.

* The rate will depend on how smooth you want the movement to be. Here is an example of a 1 Hz publish:

```python
rostopic pub -r1 /summit_xl_control/cmd_vel geometry_msgs/Twist "linear:
  x: 0.0
  y: 0.0
  z: 0.0
angular:
  x: 0.0
  y: 0.0
  z: 1.0"
```

Now , **stop it**.

```python
rostopic pub -r1 /cmd_vel geometry_msgs/Twist "linear:
  x: 0.0
  y: 0.0
  z: 0.0
angular:
  x: 0.0
  y: 0.0
  z: 0.0"
```

## **2. Demo Navigation**


* Here you are going to see two examples of navigation: **without map and with map.**

### **2.1 Navigate without Map**

* To start the navigation without a map launch the following command:

```python
$ roslaunch sumit_xl_tools start_navigation_without_map.launch
```

This command will start the navigation with only 
   * **odometry** and
   * **lasers**,
   * generating **local and global Costmaps** on the fly by the laser readings.





* This will be used by the planner to plan the best route based on the obstacles it detects.

Now here you can select the icon **`2D Nav Goal`** to set a goal for Summit XL. 

* It will try to get there without colliding into obstacles that it detects with its Laser. 

* These obstacles are drawn in the **`localCostmap`** and **`globalcostmap`** that you can see as blue areas that it should not trespass.
